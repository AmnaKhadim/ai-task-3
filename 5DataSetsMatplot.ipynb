```python
from datasets import load_dataset
import matplotlib.pyplot as plt
```

```python
dataset1 = load_dataset('ag_news', split='train')
dataset2 = load_dataset('imdb', split='train')
dataset3 = load_dataset('yelp_review_full', split='train')
dataset4 = load_dataset('squad', split='train')
dataset5 = load_dataset('glue', 'sst2', split='train')

x = [1, 2, 3, 4, 5]
y = [len(dataset1), len(dataset2), len(dataset3), len(dataset4), len(dataset5)]
```

```python
plt.plot(x, y, marker='o')  # Add markers to the line

plt.xlabel('Dataset Number')
plt.ylabel('Number of Samples')
plt.title('Number of Samples in Different Datasets')
```

```python
plt.xticks(x, ['AG News', 'IMDB', 'Yelp', 'SQuAD', 'GLUE SST-2'])

plt.grid()  # Add a grid for better readability
plt.show()  # Show the plot
```
