```python
import matplotlib.pyplot as plt
```

```python

x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

plt.plot(x, y)

plt.xlabel('Number')
plt.ylabel('Squared Value')

plt.title('Number vs Squared Value')

plt.show()
```

```python
banks = ['Bank A', 'Bank B', 'Bank C', 'Bank D']
interest_rates = [7.5, 8.0, 7.8, 8.2]  # Percentage interest rates

plt.bar(banks, interest_rates, color='b')

plt.xlabel('Banks')
plt.ylabel('Interest Rate (%)')
plt.title('Personal Loan Interest Rates Comparison')

plt.show()
```

```python
traffic_sources = ['Organic Search', 'Direct', 'Social Media', 'Referral', 'Paid Search']
traffic_percentage = [45, 30, 10, 10, 5]

plt.pie(traffic_percentage, labels=traffic_sources, autopct='%1.1f%%', startangle=90)

plt.title('E-Commerce Website Traffic Source Analysis')

plt.axis('equal')

plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 5, 7, 10]

plt.scatter(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot Example')
plt.show()
```

```python
import numpy as np

data = np.random.rand(10, 10)

plt.imshow(data, cmap='hot', interpolation='nearest')
plt.colorbar()
plt.title('Heatmap Example')
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 40, 50]
bubble_size = [40, 70, 100, 150, 200]

plt.scatter(x, y, s=bubble_size, alpha=1)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Bubble Chart Example')
plt.show()
```
